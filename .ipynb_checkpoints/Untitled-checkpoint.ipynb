## Numpy Array 

Numpy is based on arrays. But what is the **Arrays**?
Arrays are similiar to python list. If you remember, list was a data type in python that acts as a container of multiple elements. For example:
```python
list1 = ["George", 3.56, 2, N.A.]
```

Each list item contains both data and type information and items could be of any data type: integer, float, string,.. . The flexibility comes with a cost; if we have a large array with the same data type, like string, we keep type information for each item unnecessarily. Numpy array is a Fixed-type list and this lack of flexibilitymake it much more efficient for storing and manipulating data.



| Python List  | Numpy Arrays |
|----|---|
| Flexible in data type | Fixed type |
| Inefficient for large datasets  | Efficient |
|Access to a single element requires box/unbox the number into a python numeric object | Each item has its own storage reference | 
|cannot change the size of an array once it is created | can  create a new (bigger) integer and "replace" the old one |
| bigger in size than numpy arrays | -- |
| One-dimension | Multi-dimension |



source: [link](https://nbviewer.jupyter.org/github/jakevdp/PythonDataScienceHandbook/blob/master/notebooks/02.01-Understanding-Data-Types.ipynb)

Lets create a numpy array from a list:

````python
arr1 = np.array([1, 2.3, 6,8 , 0.01])
print(arr1)
````

And it could be multi-dimensional

````python
arr2 = np.array ([ [1,3], [5,7]  ])
arr2
````

'arr2' is like a $2\times2$ martix. If you want to find the shape of the matrix:

````python
print("arr2 dimension is: ", arr2.ndim, "and its shape is: ", arr2.shape)
````

## Creating an Array

````python
# Create an array with a  sequence Starting at 1, ending at 10, stepping by 2
np.arange( 1,10, 2)
````

````python
# Create an array filled with one
np.ones((3,4)   , dtype='float')
````

````python
# Create an array filled with zero
np.zeros( (2,2))
````

````python
# Create an identity matrix
np.eye(3)
````

````python
# Create an array filled with 10
np.full( (3,3)  , 10)
````

````python
# Create an uninitialized array of three integers
np.empty(3)
````

````python
#Create an array of five items evenly spaced between 0 and 1
np.linspace(0,1, 5)
````

````python
# Create an array of random values between 0 and 1
np.random.random((3, 3))
````

## Slicing and Indexing

To obtain a slice of an array, you may use  
```python
x[start:stop:step]
```

````python
arr4 = np.arange(0,10)
arr4
````

````python
arr4[0:3]
````

````python
arr4[0:10:2]
````

For multi-dimensional:

````python
arr5 = np.arange(12).reshape((4,3))
arr5
````

Indexing in numpy arrays is similar to the Python list indexing. If you want to get an elemnt in `arr4`, you may use address of the item location in a tuple. Note that in Python indices start from zero.

````python
# finding the item located in third row and first column
arr5[2, 0]
````

````python
# finding the item located in third row and the last column
arr5[2, -1]
````

Values can also be changed using the index notation

````python
arr5[ 0 , 0 ] =  100
arr4
````
